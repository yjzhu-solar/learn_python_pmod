## Plot 2D Image
---
In this section, we will learn how to plot 2D images using Matplotlib. We will use sample AIA images provided by Sunpy to demonstrate how to visualize solar data effectively. We will also explore different colormaps and normalization techniques to enhance the visibility of features in the images.

First let us load the AIA images into Sunpy Maps (see more details in the [Sunpy Map Operation notebook](sunpy_map_operation.ipynb)) and extract the image arrays for plotting.

```python
smap_171 = sunpy.map.Map(AIA_171_IMAGE)
smap_193 = sunpy.map.Map(AIA_193_IMAGE)

data_171 = smap_171.data.copy() # get the image data array from the Sunpy Map object
data_193 = smap_193.data.copy()
```

### Visualize the 2D Image
---
Now we can visualize the 2D images using Matplotlib's `imshow` function. We can either create a new figure and axis, and plot the image data over the axis using `ax.imshow()`, or directly use the `plt.imshow()` function.

```python
fig, ax = plt.subplots(layout="constrained")
ax.imshow(data_171,
          origin="lower", # <-- this is important to set the origin of the image to the lower left corner, which is the standard for solar images
          cmap="sdoaia171", # <-- this is a colormap specifically designed for AIA 171 images, provided by Sunpy. You should import the sunpy.map package to use it.
          vmin = np.nanpercentile(data_171, 1), # <-- this is to set the minimum value of the colormap to the 1st percentile of the data.
          vmax = np.nanpercentile(data_171, 99) # <-- this is to set the maximum value of the colormap to the 99th percentile of the data.
          )
```

#### No `origin=lower` (default)
---
This is the default behavior of `imshow`, where the origin of the image is set to the upper left corner. This is not the standard for solar images, and the image will be flipped vertically.

```python
fig, ax = plt.subplots(layout="constrained")
ax.imshow(data_171,
          # origin="lower", # <-- this is important to set the origin of the image to the lower left corner, which is the standard for solar images
          cmap="sdoaia171", # <-- this is a colormap specifically designed for AIA 171 images, provided by Sunpy. You should import the sunpy.map package to use it.
          vmin = np.nanpercentile(data_171, 1), # <-- this is to set the minimum value of the colormap to the 1st percentile of the data.
          vmax = np.nanpercentile(data_171, 99) # <-- this is to set the maximum value of the colormap to the 99th percentile of the data.
          )
```
